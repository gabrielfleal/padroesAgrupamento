```python
# Importa as bibliotecas necessáriass
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import string
from nltk.stem import PorterStemmer
import numpy as np
```

```python
d = open("corpora.txt").read().splitlines()
len(d)
```

```python
stopWords = set(stopwords.words('english'))
punctuations = list(string.punctuation)

#Antes tinhamos uma lista com todas palavras, mas elas se referiam a um unico documento
#Agora criamos uma lista com as listas de palavras de cada documento
corpusFiltered = []
ps = PorterStemmer()
stopWordsFilter = ["and/or","'s", "»", "â€", "ï", "“"];
for i in range(0,len(d)):
    data = d[i]
    words = word_tokenize(data)
    wordsFiltered = []
    for w in words:
        if w.lower() not in stopWords and w.lower() not in stopWordsFilter and w not in punctuations and not w.isdigit() and "â€" not in w:
            radical = ps.stem(w.lower())
            #print(radical)
            #print(w)
            wordsFiltered.append(radical)    
    #para cada documento, atualizo a lista do corpus 
    len(wordsFiltered)
    corpusFiltered.append(wordsFiltered)
print(corpusFiltered)
```

```python
for i in range(0,15):
    print("-----Document",i+1,"-----")
    print(corpusFiltered[i])
    print("--------")
```

```python
AF = {} #frequencia absoluta
TF = {} #frequencia ponderada
DF = {} #document frequency
IDF = {} #inverse document frequency
TF_IDF = {} #tfidf

terms = []
dt = {} #relacao documento termos
```

```python
#Atualizamos todos os termos, eliminando repetições
#Atualizamos a relação documento termos
terms = []
for i in range(0,15):
    terms = terms + corpusFiltered[i]    
    dt[i] = corpusFiltered[i]
terms = set(terms)
terms = list(terms)
terms.sort()
print(terms)
```

```python
#Computa a frequencia absoluta dos termos]

for i in dt: 
    for j in terms:
        AF[j,i] = dt[i].count(j)
```

```python
print(AF)
```

```python
for i in terms:
    DF[i]=0
    for j in dt:
        if i in str(dt[j]):
            DF[i]=DF[i]+1
            continue
            
            
            
```

```python
print(DF)
```

```python
import math
for i in dt:
        for j in terms:
            if AF[j,i]>0:
                TF[j,i]=1 + math.log10(AF[j,i])
                IDF[j]=math.log10((9/DF[j]))  
            else:
                TF[j,i]=0
                IDF[j,i]=0
             
```

```python
print(TF)
```

```python
print(IDF)
```

```python

for i in dt:
    for j in terms:
        TF_IDF[j,i] = TF[j,i] * IDF[j]
#colula soma para avaliar termos mais relevantes        
for j in terms:
    TF_IDF[j,i+1] = 0;
    for i in dt:
        TF_IDF[j,i+1] = TF[j,i] + TF_IDF[j,i+1]
        
            
```

```python
TF_IDF
```

```python
#cria uma lista apenas com os termos(indices) e a soma(valor do array)
arraySoma = {}
for j in terms:
    arraySoma[j]=TF_IDF[j,15]
   
```

```python
#ordena o array de soma em ordem decrescente retornando uma tupla (indice,valor)
somaOrdenada = sorted(arraySoma.items(), key=lambda kv: kv[1], reverse=True)  # Items returns a list of `(key,value)`-pairs
```
